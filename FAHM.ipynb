```python
#Can we predict the downtime rate for the next day of operation?
dfmerged2g=dfmerged2.groupby('Date.1').agg({'Batch':'count','Total errors' : 'sum'})
dfmerged2g=dfmerged2g.reset_index()
dfmerged2g.columns=['Date.1','Batch','Total errors']
dfmerged2g
```

```python
dfmerged2g.iloc[0:10].plot(
    kind='bar',
    x='Date.1',
    y='Total errors',
    title='Downtime Prediction',figsize=(12,6))
avg = dfmerged2g["Total errors"].iloc[0:50].mean()

# Ø¥Ø¶Ø§ÙØ© Ø®Ø· Ø£ÙÙ‚ÙŠ Ø¹Ù†Ø¯ Ø§Ù„Ù…ØªÙˆØ³Ø·
plt.axhline(y=avg, color='red', linestyle='--')
```
